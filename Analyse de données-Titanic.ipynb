```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset Titanic chargé directement depuis internet
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df
```

```python
# Taux de survie par genre
df.groupby('Sex')['Survived'].mean().plot(kind='bar', color=['lightblue', 'pink'])
plt.title('Taux de survie par genre')
plt.xlabel('Genre')
plt.ylabel('Taux de survie')
plt.tight_layout()
plt.show()
```

```python
# Taux de survie par classe
df.groupby('Pclass')['Survived'].mean().plot(kind='bar', color=['gold', 'silver', 'brown'])
plt.title('Taux de survie par classe')
plt.xlabel('Classe (1=Première, 2=Seconde, 3=Troisième)')
plt.ylabel('Taux de survie')
plt.tight_layout()
plt.show()
```

```python
# Distribution des âges
df['Age'].dropna().plot(kind='hist', bins=30, color='steelblue', edgecolor='black')
plt.title('Distribution des âges des passagers')
plt.xlabel('Âge')
plt.ylabel('Nombre de passagers')
plt.tight_layout()
plt.show()
```

```python
# Statistiques générales
print("=== RÉSUMÉ TITANIC ===")
print(f"Nombre de passagers : {len(df)}")
print(f"Taux de survie global : {df['Survived'].mean()*100:.1f}%")
print(f"Âge moyen : {df['Age'].mean():.1f} ans")
print(f"Prix moyen du billet : {df['Fare'].mean():.2f} £")
print(f"Passagers en 1ère classe : {len(df[df['Pclass']==1])}")
print(f"Passagers en 2ème classe : {len(df[df['Pclass']==2])}")
print(f"Passagers en 3ème classe : {len(df[df['Pclass']==3])}")
```

```python
# Prix moyen du billet par classe
df.groupby('Pclass')['Fare'].mean().plot(kind='bar', color=['gold', 'silver', 'brown'])
plt.title('Prix moyen du billet par classe')
plt.xlabel('Classe')
plt.ylabel('Prix (£)')
plt.tight_layout()
plt.show()
```

```python
# Survie par tranche d'âge
df['tranche_age'] = pd.cut(df['Age'], bins=[0,12,18,35,60,80], labels=['Enfant','Ado','Adulte','Senior','Vieux'])
df.groupby('tranche_age')['Survived'].mean().plot(kind='bar', color='teal')
plt.title('Taux de survie par tranche âge')
plt.xlabel('Tranche âge')
plt.ylabel('Taux de survie')
plt.tight_layout()
plt.show()
```
